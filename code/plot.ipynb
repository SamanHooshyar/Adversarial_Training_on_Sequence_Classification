```python
import pickle
import matplotlib.pyplot as plt
with open("test_acc_adv_lstm_ep_0p01_300.pkl","rb") as f:
    z = pickle.load(f)
    
epochs = list(range(1, 301)) 

l_acc = z


plt.rcParams.update({'font.size': 15, 'font.family': 'serif', "figure.figsize" : (6,5)})

fig1, ax1 = plt.subplots()

ax1.plot(epochs, l_acc, color='red')
# ax1.set_title('Test Accuracy Basic LSTM')  
ax1.set_title('Test Accuracy Adv LSTM for best epsilon') 
ax1.set_xlabel("No. of epochs")
ax1.set_ylabel('Test Accuracy')

print("figures have been plotted")
fig1.savefig('../figures/test-acc_advlstm_best.png')     
```

```python
with open("test_acc_adv_lstm_ep_0p01.pkl","rb") as f:
    a = pickle.load(f)
    
with open("test_acc_adv_lstm_ep_0p1.pkl","rb") as f:
    b = pickle.load(f)   
    
with open("test_acc_adv_lstm_ep_1p0.pkl","rb") as f:
    c = pickle.load(f)    
    
epochs = list(range(1, 51)) 


plt.rcParams.update({'font.size': 15, 'font.family': 'serif', "figure.figsize" : (6,5)})

fig1, ax1 = plt.subplots()

ax1.plot(epochs, a, color='red', label = "epsilon_0.01")
ax1.plot(epochs, b, color='blue', label = "epsilon_0.1")
ax1.plot(epochs, c, color='green', label = "epsilon_1")
ax1.set_title('Test Accuracy Adv LSTM for different epsilons')  
ax1.set_xlabel("No. of epochs")
ax1.set_ylabel('Test Accuracy')
ax1.legend()
ax1.set_ylim(10, 100)

print("figures have been plotted")
fig1.savefig('../figures/test-acc_adv_lstm_2.png')  
```
